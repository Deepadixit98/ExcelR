## Pandas Piechart
- piechart is suitable for **descrete numerical values columnes having less no.of unique values** & **catagorical values**.

- piechart is suitable for **frequency** and **percentage values**.

- piechart is **not** suitable for **continuous numerical values**.

```python
a= [20,30,25,36]
b=['a','b','c','d']
```

```python
plt.pie(a, labels=b, autopct = '%0.0f%%')
plt.show()
```

```python
plt.pie(df['body-style'].value_counts())
```

```python
df['body-style'].unique()
```

```python
plt.pie(df['body-style'].value_counts(), labels=df['body-style'].unique())
```

- Here we can see that the values and piechart ratio does'nt matches because it is actually taking the labels as per arrray of df's unique values. 

see output of -

- because matplot runs the loop sequencially according to array values and pie chart shows it in reverse order

```python
 names = ['sedan','hatchback','wagon', 'hardtop', 'convertible']
```
